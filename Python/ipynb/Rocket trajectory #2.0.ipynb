```python
teta = 52
#gravity
g = 9.8
#cordinates
x = 0
y = 0
#initial velocity
v0 = 19.5
#time
t = 0
#frames
t_delta = 0.01
#fricton
f = 0
```

```python
import math
teta_rad = teta * math.pi / 180
#vx_before
vx_bf = v0 * math.cos(teta_rad)
#vf_before
vy_bf = v0 * math.sin(teta_rad)
y_bf = y
x_bf = x
t_bf = 0
vx = vx_bf
vy = vy_bf
```

```python
#looping
listx1 = []
listy1 = []
while y >= -0.01:
    listx.append(round(x,2))
    listy.append(round(y,2))
    print(round(t,2),round(x,2),round(y,2),round(vx,2),round(vy,2),round(teta,2))
    #new velocities and new position
    vy = vy_bf - g * t_delta - f * vy_bf * t_delta
    y = y_bf + vy * t_delta
    #new velocities and new position
    vx = vx_bf - f * vx_bf * t_delta
    x = x_bf + vx * t_delta 
    #time
    t = t_bf + t_delta
    #Angle
    teta_rad = math.atan ((y-y_bf)/(x-x_bf))
    teta = 180 * teta_rad/math.pi
    
    #update variables
    vy_bf = vy 
    y_bf = y
    vx_bf = vx
    x_bf = x
    t_bf = t 
    
```

```python
import matplotlib.pyplot as plt
plt.plot(listx, listy)
plt.plot(listx1, listy1)
plt.xlabel('Distance')
plt.ylabel('Height')
plt.show()
```
